```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df2 = df[['C1', 'C2', 'C3', 'C4', 'C5', 'C6']]
df2.head()
```

```python
weights = np.array([15, 15, 16, 19, 18, 17])
```

```python
def normalize(weights):
    max_n = sum(weights)
    weights_n = weights / max_n
    return weights_n
```

```python
weights_n = normalize(weights)
print(weights_n)
```

```python
df3 = df2.multiply(weights_n)
```

```python
df['TOTAL'] = df3.sum(axis=1)
df.head()
```

```python
def hitung_utility(df):
    d_max = df.max(axis=1)
    d_min = df.min(axis=1)
    
    res = df.subtract(d_max, axis=0).abs()
    return res
```

```python
df2.max(axis=1).head()
```

```python
df2.head()
```

```python
res = hitung_utility(df2)
res.head()
```

```python
d_max = df2.max(axis=1)
d_min = df2.min(axis=1)
d_int = d_max - d_min

d_int.head()
```

```python
df4 = res.divide(d_int, axis=0)
df4.head()
```
